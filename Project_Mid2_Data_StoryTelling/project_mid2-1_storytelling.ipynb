**1. Rename the [US dollar ] and Period\Unit: columns to something easier to type – US_dollar and Time**

    • Change the Time column to a datetime data type.
    • Sort the values by Time in ascending order.
    • Reset the index (and drop the initial index).

```python
currency = currency.rename(columns={'[US dollar ]': 'US_dollar',
                        r'Period\Unit:': 'Time'})
currency.columns
```

```python
#  Change the Time column to a datetime data type.
currency['Time'] = pd.to_datetime(currency['Time'])
currency['Time']
```

```python
#  Sort the values by Time in ascending order.
currency = currency.sort_values(by='Time', ascending= True)

# Reset the index (and drop the initial index).
currency.reset_index(inplace=True)
currency.drop(columns='index',inplace=True)
currency
```

**Isolate the Time and the US_dollar columns. Assign them to a different variable named euro_to_dollar**

```python
euro_to_dollar = currency[['Time','US_dollar']].copy()
euro_to_dollar
```

```python
euro_to_dollar['US_dollar'].value_counts(dropna=False)
```

**Convert the US_dollar column to a float data typ**

```python
euro_to_dollar = euro_to_dollar[euro_to_dollar['US_dollar'] != '-']
euro_to_dollar['US_dollar'] = euro_to_dollar['US_dollar'].astype(float)
euro_to_dollar['US_dollar'].value_counts(dropna=False)
```

If we look at the line’s shape, we see many small wiggles – rather than seeing a smooth line. The wiggles, 
however, have meaning: they are the visual representation of the daily variation in the exchange rate. The rate 
goes up and down, up and down again, day to day. The rate only shows clear upward or downward #009688ds in the 
longer run (months or years).

```python
fig, ax = plt.subplots(figsize = (9,6))
ax.plot(euro_to_dollar['Time'],euro_to_dollar['US_dollar'],linewidth = 2.0,color = 'navy')
```

Depend on our goals, we may not want to show that daily variation on our graph. If we want to hide it and show 
only the long-term #009688ds, we can use the rolling mean ( moving average).


**Calculate the mean for each day using the pd.Series.rolling().mean() method**

    • Ex: values[‘daily_values’].rolling(2).mean()
    
    • Calculate the rolling means for the US_dollar column using a moving window of 30 days. Add the rolling 
    means to a new column named rolling_mean 

```python
euro_to_dollar['rolling_mean'] = euro_to_dollar['US_dollar'].rolling(30).mean()
euro_to_dollar
```

```python
# Plot the chart agian to see whether we revome the wiggle
fig, ax = plt.subplots(figsize=(9, 6))
ax.plot(euro_to_dollar['Time'], euro_to_dollar['rolling_mean'],linewidth = 2.8,color = 'navy')
```

## Storytelling : Compare the evolution of exchange rate of Euro to the New Zealand dollar (NZD) and Canadian dollar (CAD) over a period from 2005 to 2022

**Creating a graph for multiple currencies –  compare their evolution**

```python
currency = currency.rename(columns={'[New Zealand dollar ]': 'NZD'})
currency = currency.rename(columns={'[Canadian dollar ]': 'CAD'})
```

```python
euro_to_nzd = currency[['Time','NZD']]
euro_to_cad = currency[['Time','CAD']]
```

```python
euro_to_nzd = euro_to_nzd.replace('-',np.nan)
euro_to_nzd = euro_to_nzd.dropna()
euro_to_nzd['NZD'] = euro_to_nzd['NZD'].astype(float)
euro_to_nzd['rolling_mean'] = euro_to_nzd['NZD'].rolling(30).mean()
```

```python
euro_to_cad = euro_to_cad.replace('-',np.nan)
euro_to_cad = euro_to_cad.dropna()
euro_to_cad['CAD'] = euro_to_cad['CAD'].astype(float)
euro_to_cad['rolling_mean'] = euro_to_cad['CAD'].rolling(30).mean()
```

```python
euro_to_nzd_all = euro_to_nzd.loc[euro_to_nzd["Time"].dt.year.between(2005,2022)]
euro_to_cad_all = euro_to_cad.loc[euro_to_cad["Time"].dt.year.between(2005,2022)]
euro_to_nzd_crisis = euro_to_nzd.loc[euro_to_nzd["Time"].dt.year.between(2007,2009)]
euro_to_cad_crisis = euro_to_cad.loc[euro_to_cad["Time"].dt.year.between(2007,2009)]
euro_to_nzd_pandemic = euro_to_nzd.loc[euro_to_nzd["Time"].dt.year.between(2016,2020)]
euro_to_cad_pandemic = euro_to_cad.loc[euro_to_cad["Time"].dt.year.between(2016,2020)]
```

```python
import matplotlib.pyplot as plt

# Create the plot
fig, ax = plt.subplots(figsize=(18, 6))
ax.set_ylim(1.1, 2.75)
ax.set_yticks([1.5, 1.75, 2, 2.25, 2.5, 2.75])
ax.set_yticklabels(['1.5', '1.75', '2', '2.25', '2.5', '2.75'])
# ax.grid(alpha=0.3)
# ax.set_xlim(xmin=euro_to_cad[euro_to_cad_all["Time"].dt.year >= 2006)

# Plot the NZD exchange rate data
ax.plot(euro_to_nzd_all['Time'],
        euro_to_nzd_all['rolling_mean'], color='#009688')
ax.plot(euro_to_cad_all['Time'],
        euro_to_cad_all['rolling_mean'], color='#008000')
ax.spines[['top', 'right']].set_visible(False)
ax.legend(['New Zealand', 'Canada'])
# Set labels and title
ax.set_xlabel('Year', fontsize=15, weight='bold')
ax.set_ylabel('Exchange Rate', fontsize=15, weight='bold')
ax.set_title('Exchange Rate of EURO to NZD and CAD',
             size=25, color='teal', weight='bold', y=1.07)

# Fill each stage of exchange rate
ax.fill_between(euro_to_nzd_all['Time'], euro_to_nzd_all['rolling_mean'], where=(euro_to_nzd_all["Time"].dt.year >= 2006) & (euro_to_nzd_all["Time"].dt.year < 2009), color='#009688', alpha=0.2)
ax.fill_between(euro_to_cad_all['Time'], euro_to_cad_all['rolling_mean'], where=(euro_to_cad_all["Time"].dt.year >= 2006) & (euro_to_cad_all["Time"].dt.year < 2009), color='#B9E5CD')

ax.fill_between(euro_to_nzd_all['Time'], euro_to_nzd_all['rolling_mean'], where=(euro_to_nzd_all["Time"].dt.year >= 2020) & (euro_to_nzd_all["Time"].dt.year < 2023), color='#009688', alpha=0.2)
ax.fill_between(euro_to_cad_all['Time'], euro_to_cad_all['rolling_mean'], where=(euro_to_cad_all["Time"].dt.year >= 2020) & (euro_to_cad_all["Time"].dt.year < 2023), color='#B9E5CD')


highest_value = euro_to_cad_crisis["rolling_mean"].max()
highest_value_time = euro_to_cad_crisis.loc[euro_to_cad_crisis["rolling_mean"]== highest_value, "Time"].values[0]
ax.annotate(f"CAD Peak: {highest_value:.2f}", xy=(highest_value_time, highest_value),
            xytext=(14500, highest_value + 0.5), arrowprops=dict(arrowstyle="-"), weight='bold', fontsize=12)

highest_value = euro_to_nzd_crisis["rolling_mean"].max()
highest_value_time = euro_to_nzd_crisis.loc[euro_to_nzd_crisis["rolling_mean"] == highest_value, "Time"].values[0]
ax.annotate(f"NZD Peak: {highest_value:.2f}", xy=(highest_value_time, highest_value), xytext=(14500, 2.6), arrowprops=dict(arrowstyle="-"), weight='bold', fontsize=12)


ax.text(18500, 1.2, 'PANDEMIC', fontsize=17, color='#003200')
ax.text(13500, 1.2, 'CRISIS', fontsize=17, color='#003200')

plt.axvline(pd.Timestamp("2008-12-31"), color='gray',
            linewidth=1.8, linestyle='--')
plt.axvline(pd.Timestamp("2020-1-1"), color='gray',
            linewidth=1.8, linestyle='--')
# Show the plot
plt.show()
```

**Conclusion:   Both the NZD and CAD weakened against the EURO during the financial crisis and the COVID-19 pandemic.**
