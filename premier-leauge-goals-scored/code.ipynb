```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv(r'D:\NEU\NÄƒm 3\DATA PREP\Data\premier-leauage-goal-scores.csv')
data
```

```python
data = data.sort_values(by='Goals', ascending=True)
data['total'] = data['Goals'].max()
data = data.reset_index()
data['index'] = list(range(20))
data
```

```python
fig, ax = plt.subplots(figsize=(9, 10))
ax.barh(data['index'], data['total'], height=0.8, color='#F5F5F5', alpha=0.7)
ax.barh(data['index'], data['Goals'], height=0.8, color='#151D6F')
ax.spines[['left', 'right', 'top', 'bottom']].set_visible(False)
ax.set_yticks(list(range(20)))

# annotate the number of goals of each club
for i in range(len(data)):
    ax.annotate(xy=(0.2, data['index'][i]-0.15),
                text='{}'.format(data['Goals'][i]), color='#F5F5F5', size=11.5)


# add title and subtitle
ax.text(x=5, y=21.1, s='Goals scored by forwards',
        weight='bold', color='black', size=17.0, alpha=0.7)
ax.text(x=6, y=20.3, s='Premier Leauge 2023 - 2024', color='#B2B2B2', size=13)

# modify ticks
ax.set_yticklabels(list(data['Club']))
ax.set_xticks([])
ax.tick_params(left=False, bottom=False)
ax.set_ylim(ymin=-0.5)

# highlight MU position
ax.get_yticklabels()[0].set_color('maroon')
```
